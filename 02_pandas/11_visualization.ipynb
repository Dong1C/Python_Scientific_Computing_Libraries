## 11. Visualization

- **Plotting**: `plot()`, integrated with Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrame
data = {'A': [1, 2, 3, 4, 5],
        'B': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)

# Line plot from DataFrame
df.plot()
plt.title('Line Plot')
plt.show()

# Line plot from Series
df['A'].plot()
plt.title('Line Plot from Series')
plt.show()
```

```python
# Bar plot from DataFrame
df.plot(kind='bar')
plt.title('Bar Plot')
plt.show()

# Bar plot from Series
df['A'].plot(kind='bar')
plt.title('Bar Plot from Series')
plt.show()
```

```python
# Horizontal bar plot from DataFrame
df.plot(kind='barh')
plt.title('Horizontal Bar Plot')
plt.show()

# Horizontal bar plot from Series
df['A'].plot(kind='barh')
plt.title('Horizontal Bar Plot from Series')
plt.show()
```

```python
# Histogram from DataFrame
df.plot(kind='hist', alpha=0.7)
plt.title('Histogram')
plt.show()

# Histogram from Series
df['A'].plot(kind='hist', alpha=0.7)
plt.title('Histogram from Series')
plt.show()
```

```python
# Box plot from DataFrame
df.plot(kind='box')
plt.title('Box Plot')
plt.show()

# Box plot from Series
df['A'].plot(kind='box')
plt.title('Box Plot from Series')
plt.show()
```

```python
# Scatter plot from DataFrame
df.plot(kind='scatter', x='A', y='B')
plt.title('Scatter Plot')
plt.show()
```

```python
# Hexbin plot from DataFrame
df.plot(kind='hexbin', x='A', y='B', gridsize=10)
plt.title('Hexbin Plot')
plt.show()
```

```python
# Pie chart from Series
df['A'].plot(kind='pie', autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
# Area plot from DataFrame
df.plot(kind='area', alpha=0.5)
plt.title('Area Plot')
plt.show()

# Area plot from Series
df['A'].plot(kind='area', alpha=0.5)
plt.title('Area Plot from Series')
plt.show()
```

```python
# KDE plot from Series
df['A'].plot(kind='kde')
plt.title('KDE Plot from Series')
plt.show()
```

```python
# Customizing plots
ax = df.plot(kind='line', title='Customized Plot')
ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label')
plt.show()
```
