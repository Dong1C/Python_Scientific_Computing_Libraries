# CrossTab and Accumulation

```python
df_stock['cate'] = [random.randint(1, 3) for i in range(df_stock.shape[0])]
df_stock
```

```python
type(df_stock.groupby("cate"))
```

```python
class a: 
    def __type__(self):
        return type(self)
    
    def __call__(self):
        return self.__type__()
    
eval(''.join(['a', *['()' for _ in range(100)]]))
```

```python
dsg = df_stock.groupby("cate")
agg_1 = dsg.agg(lambda x: x.sum() / x.count())
agg_1.columns.name = 'Company'
```

```python
# stack and unstack to change
# set the name of columns
# groupby([$1, $2]) then unstack will bring the $2 to the column name...
agg_1.reset_index(drop=False)
agg_1.unstack().index
```
